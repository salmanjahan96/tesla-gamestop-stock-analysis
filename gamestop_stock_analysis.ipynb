# GameStop Stock Analysis

```
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```
gme = yf.Ticker('GME')
gme_data = gme.history(period='max')
gme_data.reset_index(inplace=True)
gme_data.head()
```

```
gme_data = gme_data[['Date','Open','High','Low','Close','Volume']]
gme_data.to_csv('gme_stock_data.csv', index=False)
print(gme_data.tail())
```

```
plt.figure(figsize=(12,5))
plt.plot(gme_data['Date'], gme_data['Close'])
plt.title('GameStop Historical Closing Price')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```
print('GameStop is highly volatile with sharp spikes during meme-stock events.')
```
