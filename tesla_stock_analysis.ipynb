# Tesla Stock Analysis

```
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```
tesla = yf.Ticker('TSLA')
tesla_data = tesla.history(period='max')
tesla_data.reset_index(inplace=True)
tesla_data.head()
```

```
tesla_data = tesla_data[['Date','Open','High','Low','Close','Volume']]
tesla_data.to_csv('tesla_stock_data.csv', index=False)
print(tesla_data.tail())
```

```
plt.figure(figsize=(12,5))
plt.plot(tesla_data['Date'], tesla_data['Close'])
plt.title('Tesla Historical Closing Price')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```
print('Tesla shows strong long-term growth with occasional volatility.')
```
